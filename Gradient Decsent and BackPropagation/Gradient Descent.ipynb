```python
import numpy as np
import matplotlib.pyplot as plt
```

# 1D Gradient Descent

```python
global a, alpha
a =3
alpha = 0.01
```

```python
def dev_E(x):
    return 2*(x-a)

def E(x):
    return (x-a)**2

def Run_Gradient_Descent(initial_value = 12):
    x = initial_value
    x_val = []
    for i in range(200):
        x = x - alpha*dev_E(x)
        x_val.append(x)
        
    return x, x_val
```

```python
x, x_val = Run_Gradient_Descent(12)
x
```

```python
plt.figure(figsize=(10,5))
x = np.arange(-4,13,0.1)
plt.plot(x,E(x))
plt.scatter(x_val, E(np.array(x_val)), lw = 4, c = [0.1*i for i in range(len(x_val))])
plt.colorbar()
plt.axis('off')
```

# 2D Gradient Descent

### First Example

```python
def E(x1,x2):
    return (x1-a)**2 + (x2 - a)**2

def Grad_E(x):
    return 2 * (x-a)

def update_param(x):
    gradian = Grad_E(x)
    x = x - alpha*gradian
    return x

def Run_GD(x ,iteration = 150):
    x_val = []
    for i in range(iteration):
        x = update_param(x)
        x_val.append(x)
    return x, np.array(x_val)
```

```python
x = np.array([14,13])
update_param(x)
```

```python
x, x_val = Run_GD(x)
```

```python
nx, ny = (50,50)
x = np.linspace(-5,15,nx)
y = np.linspace(-5,15,ny)
```

```python
xv, yv = np.meshgrid(x,y)
```

```python
plt.contourf(xv,yv,E(xv,yv), levels=10)
plt.contour(xv,yv,E(xv,yv), levels=10,colors='k')
plt.scatter(x_val[:,0],x_val[:,1],c=[i for i in range(len(x_val[:,0]))], cmap='jet',
            s=50*np.array([np.exp(-0.05*i) for i in range(len(x_val[:,0]))])[::1])
```
